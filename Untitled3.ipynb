```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/Data Science Job Salaries.csv')
df.head()
```

```python
if 'unnamed: 0' in df.columns:
  df.drop(["Unnamed: 0"], axis=1)
else:
  print("column 'Unnamed: 0' is not found in the Dataframe")
```

```python
df.columns
```

```python
print(df.isnull().sum())
```

```python
print(df["experience_level"].value_counts(), df["employment_type"].value_counts(), df["company_size"].value_counts().head(10))
```

```python
df["job_title"].value_counts().head()
```

```python
# Simplify job titles
def simplify_job_title(title):
    title = title.lower()
    if 'scientist' in title:
        return 'Data Scientist'
    elif 'engineer' in title:
        return 'Data Engineer'
    elif 'analyst' in title:
        return 'Data Analyst'
    elif 'manager' in title:
        return 'Data Manager'
    elif 'director' in title:
        return 'Data Director'
    else:
        return 'Other'

df['simplified_job_title'] = df['job_title'].apply(simplify_job_title)
df.head()
```

```python
# Calculate descriptive Stats for Salary
salary_stats = df["salary_in_usd"].agg(['mean','median','std'])
print("Descriptive stats for Salary:\n", salary_stats)
```

```python
# Group by job title and calculate mean salary
job_title_salary = df.groupby('job_title')['salary_in_usd'].mean().sort_values(ascending=False).head(10)
print("\nMean Salary by Job Title:\n", job_title_salary)
```

```python
# Histogram of salary Distribution
plt.figure(figsize=(8,6))
plt.hist(df['salary_in_usd'], color='skyblue', bins=20, edgecolor='black')
plt.xlabel('salary_in_usd')
plt.ylabel('frequncy')
plt.title('Distribution of Salary')
plt.show()
```

**Salary Distribution:**

The mean salary is approximately  110,610, with a median of 100,000 and a standard deviation of $72,281, indicating a wide salary range.

Few people earn over $300,000

```python
# 3. Bar chart of average salary by simplified job title
plt.figure(figsize=(8, 6))
df.groupby('simplified_job_title')['salary_in_usd'].mean().plot(kind='bar', color='salmon')
plt.xlabel('Simplified Job Title')
plt.ylabel('Average Salary in USD')
plt.title('Average Salary by Job Title')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

**Job Title Impact:**

 Data Director roles command the highest average salaries, followed by other roles.

```python
top_titles = df['job_title'].value_counts()[:3]
top_titles
```

```python
plt.figure(figsize=(15,5))
plt.title('Top 3 Job Titles by Experience Level', fontsize=15, fontweight='bold')
ax=sns.countplot(x='job_title', hue='experience_level', data=df,
              order=df['job_title'].value_counts().iloc[:3].index)
for i in ax.containers:
    ax.bar_label(i)
plt.ylabel('Count', fontsize=10, fontweight='bold')
plt.xlabel('Job Title', fontsize=10, fontweight='bold')
plt.show()
```

```python
# Mean Salary of Employee with Different Experience level
experience_level_salary = df.groupby('experience_level')['salary_in_usd'].mean().sort_values()
experience_level_salary
```

```python
 plt.figure(figsize=(8,6))
 plt.subplot(1, 1, 1)
 ax = sns.barplot(x=experience_level_salary.index, y=experience_level_salary, palette='spring')
 ax.set_title('Mean Salary Vs Experience Level',
 fontdict={'fontsize': 10})
```

**Experience level vs Salary:**

We see that data scientists with the experience level of Executive have the highest mean salary, about 200,000 annualy, and those with Entry level have the lowest mean salary, about 60,000.

```python
plt.figure(figsize=(10,5))
sns.kdeplot(data=df, x='salary_in_usd', hue='work_year', fill=True)
plt.title("Distribution of salary by work year", fontsize=16)
plt.xlabel("Salary (in USD)", fontsize=14)
plt.ylabel("Density", fontsize=14)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.show()
```

The density plot shows an increase in the number of Data Science jobs and salary for each year.

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='employment_type',y='salary_in_usd',data=df, hue='employment_type',palette='Set2')
plt.xlabel('employment level')
plt.ylabel('salary')
plt.title('Employment type vs Salary')
plt.show()
```

 **Employment Type VS Salary:**

 We see that data scientists with an employment type of contract have the highest mean salary, about 180,000 and those who work part-time have the lowest mean salary, about 30,000 annually.

```python
#Mean Salary of Employee by different Company size
mean_s_cmp_size =df.groupby('company_size')['salary'].mean().sort_values()
mean_s_cmp_size
```

```python
plt.figure(figsize=(14, 7))
sns.set_palette('spring')
plt.subplot(1, 2, 1)
ax = sns.barplot(x=mean_s_cmp_size.index, y=mean_s_cmp_size)
ax.set_title('Mean Salary VS Company Size',fontdict={'fontsize': 16})
```

**Company Size Influence:**

Medium and large companies tend to offer higher average salaries compared to small companies. The combination of not having medium or large company size, but having small company size, correlates with a lower average salary.

```python
# 4. Scatter plot of salary vs. remote ratio, colored by experience level
plt.figure(figsize=(10, 6))
sns.scatterplot(x='remote_ratio', y='salary_in_usd', hue='experience_level', data=df, palette="viridis")
plt.xlabel('Remote Ratio')
plt.ylabel('Salary in USD')
plt.title('Salary vs. Remote Ratio (colored by experience level EX)')
plt.show()
```

**Remote Work Correlation:**

Fully remote positions (100% remote ratio) show the highest average salaries, followed by on-site positions (0%).
